```python
import pandas as pd
import re

def convert_price(price):
    if isinstance(price, str):
        price = price.replace("PKR", "").replace(",", "").strip().lower()

        if "crore" in price:
            return float(price.replace("crore", "").strip()) * 10000000
        if "million" in price:
            return float(price.replace("million", "").strip()) * 1000000
        if "lac" in price or "lakh" in price:
            return float(price.replace("lac", "").replace("lakh", "").strip()) * 100000
        if price.isdigit():
            return float(price)

    return None

df["Price_clean"] = df["Price"].apply(convert_price)
```

```python
df[["Price", "Price_clean"]].head(10)
```

```python
import re

def convert_area(area):
    if isinstance(area, str):
        a = area.lower().replace(",", "").strip()
        
        # Kanal → Marla
        if "kanal" in a:
            num = float(re.findall(r"\d+\.?\d*", a)[0])
            return num * 20
        
        # Marla → Marla
        if "marla" in a:
            num = float(re.findall(r"\d+\.?\d*", a)[0])
            return num
        
        # Sqft → Marla (1 Marla = 225 sq.ft.)
        if "sq" in a or "ft" in a:
            num = float(re.findall(r"\d+\.?\d*", a)[0])
            return num / 225
        
    return None

df["Area_marla"] = df["Area"].apply(convert_area)
```

```python
df[["Area", "Area_marla"]].head(10)
```

```python
df["Price_per_marla"] = df["Price_clean"] / df["Area_marla"]
```

```python
df[["Title", "Location", "Area_marla", "Price_clean", "Price_per_marla"]].head(10)
```

```python
def extract_society(location):
    if isinstance(location, str):
        # Split by commas
        parts = location.split(",")
        # Take the first part
        return parts[0].strip()
    return location

df["Society"] = df["Location"].apply(extract_society)
```

```python
df[["Location", "Society"]].head(10)
```

```python
df["Date_Posted_clean"] = pd.to_datetime(df["Date Posted"], errors='coerce')
```

```python
df[["Date Posted", "Date_Posted_clean"]].head(10)
```

```python
avg_price_society = df.groupby("Society")["Price_per_marla"].mean().sort_values(ascending=False)
print(avg_price_society.head(10))
```

```python
df = df[df["Area_marla"] > 0]
```

```python
avg_price_society = df.groupby("Society")["Price_per_marla"].mean().sort_values(ascending=False)
print(avg_price_society.head(10))
```

```python
def format_price(x):
    if x >= 10000000:  # 1 Crore
        return f"{x/10000000:.2f} Cr"
    elif x >= 100000:   # 1 Lakh
        return f"{x/100000:.2f} Lakh"
    else:
        return f"{x:.0f}"

# Apply formatting to average price per Marla
avg_price_society_formatted = avg_price_society.apply(format_price)
print(avg_price_society_formatted.head(10))
```

```python
df["Price_per_marla_formatted"] = df["Price_per_marla"].apply(format_price)
```

```python
df_powerbi = df[[
    "Title", 
    "Society", 
    "Area_marla", 
    "Price_clean", 
    "Price_per_marla", 
    "Price_per_marla_formatted",
    "Bedrooms", 
    "Bathrooms", 
    "Date_Posted_clean"
]]
```
